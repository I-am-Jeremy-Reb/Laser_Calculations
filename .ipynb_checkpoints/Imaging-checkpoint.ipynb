```python
def Image_Distance(f,u):
    print(f'Focal length is {f} and object distance is {u}')
    ##image is
    v= (f*u)/(u-f)
    print("The image distance is ", v )
    return v

def magnification(u,v):
    print(f'Object distance is {u} and image distance is {v}')
    return -(v/u)
```
